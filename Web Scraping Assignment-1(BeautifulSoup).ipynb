```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def scrap(URL):
    page= requests.get(URL)
    if (page.status_code == 200):
        soup= BeautifulSoup(page.content)
        
        period= soup.find_all('p', class_= 'period-name')
        periods= []
        for i in period:
            periods.append(i.text.replace('\n', ''))
        periods1=[]
        periods2=[]
        for i in periods:
            if  periods.index(i)%2 == 0:
                periods1.append(i)
            else:
                periods2.append(i)
        periods1.extend(periods2)
        
        
        discre= soup.find_all('p', class_= 'short-desc')
        discres= []
        for i in discre:
            discres.append(i.text.replace('\n', ''))
        discres1=[]
        discres2=[]
        for i in discres:
            if  discres.index(i)%2 == 0:
                discres1.append(i)
            else:
                discres2.append(i)
        discres1.extend(discres2)   
        
            
        high_temp= soup.find_all('p', class_= 'temp temp-high')  
        high_temps= []
        for i in high_temp:
            high_temps.append(i.text.replace('\n', ''))
        low_temp= soup.find_all('p', class_= 'temp temp-low')
        low_temps= []
        for i in low_temp:
            low_temps.append(i.text.replace('\n', ''))
        high_temps.extend(low_temps)
        t1= []
        t2= []
        for i in high_temps:
            t1.append(i.split(':')[0])
        for i in high_temps:
            t2.append(i.split(':')[1])
        
        
   
        
            
        print(periods1)
        print('\n\n')
        print(discres1)
        print('\n\n')
        print(t1)
        print('\n\n')
        print(t2)
        

        weather_forcast= pd.DataFrame({})
        weather_forcast['Periods']= periods1
        weather_forcast['Discription']= discres1
        weather_forcast['Temp_Discription']= t1
        weather_forcast['Temp']= t2
        return weather_forcast, weather_forcast.to_csv('weather_forcast.csv') 
        
        
   
    else:
        print('Data Scraping is not Legal from given website')
scrap('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YJfB9LUzaUk')
```
