1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver= webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
numbers= []
names= []
artist = []
Views = []
date = []

for i in range(1,31):
    for i in driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr["+str+"]/td[1]'):
        numbers.append(i.text.replace("\n", ' '))
     
    for i in driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr["+str+"]/td[2]'):
        names.append(i.text.replace("\n", ' ').split("[")[0])
    
    for i in driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr["+str+"]/td[3]'):
        artist.append(i.text.replace("\n", ' '))
        
    for i in driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr["+str+"]/td[4]'):
        Views.append(i.text.replace("\n", ' '))
        
    for i in driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr["+str+"]/td[5]'):
        date.append(i.text.replace("\n", ' '))
        
        
dictionnary = {
   'Rank': numbers[:30], 
   'Name': names[:30], 
   'Artist':artist[:30], 
   'Views':Views[:30],
    'Upload Date': date[:30]
}
top_videos = pd.DataFrame(data=dictionnary) 
top_videos   
```

C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
driver= webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.bcci.tv/')
```

```python
driver.get(driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a').get_attribute('href'))
```

```python
titles= []
series= []
places= []
dates= []
months= []
new= []
times= []

for i in driver.find_elements_by_xpath("//span[@class= 'u-unskewed-text fixture__tournament-label u-truncated']"):
    titles.append(i.text.replace("\n", ""))

for i in driver.find_elements_by_xpath("//strong[@class= 'fixture__name fixture__name--with-margin']"):
    series.append(i.text.replace("\n", ""))

for i in driver.find_elements_by_xpath("//p[@class= 'fixture__additional-info']/span"):
    places.append(i.text.replace("\n", ""))
    
for i in driver.find_elements_by_xpath("//span[@class= 'fixture__date']"):
    dates.append(i.text.replace("\n", ""))

for i in driver.find_elements_by_xpath("//span[@class= 'fixture__month']"):
    months.append(i.text.replace("\n", ""))
    
for i in range(len(dates)):
    new.append(dates[i]+months[i])

for i in driver.find_elements_by_xpath("//span[@class= 'fixture__time']"):
    times.append(i.text.replace("\n", ""))

    
dictionnary = {
   'Match Title': titles, 
   'Series': series, 
   'Place':places, 
   'Date':new,
    'Time': times
}
bcci = pd.DataFrame(data=dictionnary) 
bcci
```

```python
name= []
data= []
task=[]
attribute_type=[]
instances= []
attribute=[]
year=[]

for i in range(1):
    for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str+"]/td[1]/table/tbody/tr/td[2]/p/b/a'):
        name.append(i.text.replace("\n", ' '))
    for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str+"]/td[2]/p'):
        data.append(i.text.replace("\n", ' '))
    for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str+"]/td[3]/p'):
        task.append(i.text.replace("\n", ' '))
    for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str+"]/td[4]/p'):
        attribute_type.append(i.text.replace("\n", ' '))
    for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str+"]/td[5]/p'):
        instances.append(i.text.replace("\n", ' '))
    for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str+"]/td[6]/p'):
        attribute.append(i.text.replace("\n", ' '))
    for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str+"]/td[7]/p'):
        year.append(i.text.replace("\n", ' '))
        
dictionnary = {
   'Dataset name': name, 
   'Data Type': data[1:],
   'Task': task[1:], 
   'Attribute type': attribute_type[1:],
   'No. of instances': instances[1:], 
    'No. of attribute': attribute[1:],
   'Year': year[1:]}

github = pd.DataFrame(data=dictionnary)
github
```
